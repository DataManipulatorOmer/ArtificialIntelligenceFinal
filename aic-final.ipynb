**Loader Function**

```python
def loaderFunction(imagePaths, idealSize=(224, 224)):
    images = []
    errorImages = []

    for path in imagePaths:
        image = cv2.imread(path)

        if image is None:
            print(f"Error loading image: {path}")
            errorImages.append(path)
            continue

        # Resize the image to the target size
        image = cv2.resize(image, idealSize)

        images.append(image)

    if errorImages:
        print("\nProblematic images:")
        for img_path in errorImages:
            print(img_path)

    return np.array(images)
```

**Model VGG19**

```python
from keras.applications.vgg16 import VGG16
from keras.applications.vgg19 import VGG19
from keras.models import Sequential, Model

from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import LearningRateScheduler
import numpy as np
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input
from keras.optimizers import Adam
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
X_train = loaderFunction(trainingImagesPaths)
y_train = to_categorical(trainDataSet['Label'].map({'Normal': 0, 'Mitosis': 1}))
img_input = Input(shape=(224, 224, 3))
base_model = VGG19(weights=None, include_top=False, input_tensor=img_input)
base_model = VGG19(weights='imagenet', include_top=False, input_tensor=img_input)
def lr_schedule(epoch, initial_lr=0.001):
    lr = initial_lr * np.exp(-0.1 * epoch)
    return lr

# Use the LearningRateScheduler callback
lr_scheduler = LearningRateScheduler(lambda epoch: lr_schedule(epoch))

# Freeze the layers of the VGG19 base
for layer in base_model.layers:
    layer.trainable = False

# Build your model on top of VGG19
model = Sequential()
model.add(base_model)
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(2, activation='softmax'))

# Compile the model
optimizer = Adam(lr=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Split the data
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2)

# Train the model
epochs = 50
batch_size = 32
validation_split = 0.2
model.fit(X_train, y_train, epochs=epochs, batch_size=32, validation_split=0.2, callbacks=[lr_scheduler])

# Evaluate the model on the validation set
evaluation_result = model.evaluate(X_val, y_val)
print("\nEvaluation Results:")
print(f"Validation Loss: {evaluation_result[0]}")
print(f"Validation Accuracy: {evaluation_result[1]}")
```

**Model CNN WITH DIFF ACTIVATION FUNCTION Pt2**

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense


X_train = loaderFunction(trainingImagesPaths)
y_train = to_categorical(trainDataSet['Label'].map({'Normal': 0, 'Mitosis': 1}))
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(2, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
# Assuming X_train and y_train are your input and target data
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2)
evaluationResult = model.evaluate(X_val, y_val)
print("\nEvaluation Results:")
print(f"Validation Loss: {evaluationResult[0]}")
print(f"Validation Accuracy: {evaluationResult[1]}")
```

**Plot Function**

```python
def plotFunction(history):
    # Plot training & validation accuracy values
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()

    # Plot training & validation loss values
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()
```
